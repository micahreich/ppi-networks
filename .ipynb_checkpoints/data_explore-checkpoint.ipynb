```python
import json
import xmltodict
import pprint
import matplotlib
```

```python
# convert ecoli text data to lists

with open("data/ecoli/ecoli.protein.info.v11.5.txt") as file:
    lines = []
    for line in file:
        lines.append(line.split()[0:2])
    protein_names = lines
file.close()

with open("data/ecoli/ecoli.protein.links.v11.5.txt") as file:
    lines = []
    for line in file:
        lines.append(line.split())
    protein_links = lines
file.close()

protein_links = protein_links[1:]
```

```python
print(protein_names)
```

```python
# gather proteins into a dictionary for adjacency list

protein_links_dict = {}

for link in protein_links:
    protein_1, protein_2, score = link
    
    if protein_1 not in protein_links_dict: 
        protein_links_dict[protein_1] = [protein_2]
    else:
        protein_links_dict[protein_1].append(protein_2)

# gather proteins into dictionary to identify protein by name from ID
```

```python
# quick test to find the most connected protein

longest_key_len = 0
longest_key = None

for i in protein_links_dict:
    if len(protein_links_dict[i]) > longest_key_len:
        longest_key = i
```
